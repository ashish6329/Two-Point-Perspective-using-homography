```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import image
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img_a=image.imread('/content/a.png')
img_b=image.imread('/content/b.png')
img_a = cv2.cvtColor(img_a, cv2.COLOR_BGRA2BGR)
imgplot = plt.imshow(img_a)
plt.show()
img_b = cv2.cvtColor(img_b, cv2.COLOR_BGRA2BGR)
imgplot1=plt.imshow(img_b)
plt.show()
```

```python
size = img_a.shape
h1,w1 = img_a.shape[:2]
pts_source=np.array([[0,0],[w1,0],[w1,h1],[0,h1]],dtype=float)
pts_dst1=np.array([[100,120], [400,60], [400,370],[100,300]])
pts_dst2=np.array([[450,60],[750,150],[750,300],[450,370]])
height,width=img_b.shape[:2]
h1, mask=cv2.findHomography(pts_source, pts_dst1, cv2.RANSAC,5.0)
temp1=cv2.warpPerspective(img_a, h1, (width,height))
h2, mask=cv2.findHomography(pts_source, pts_dst2, cv2.RANSAC,5.0)


temp2=cv2.warpPerspective(img_a, h2, (width,height))
dst=cv2.addWeighted(img_b,1,temp1,1,0)
final_image = 255*dst.astype(np.uint32)
plt.imshow(final_image)
plt.show()
```

```python
dst1=cv2.addWeighted(img_b,1,temp2,1,0)
final_image1 = 255*dst1.astype(np.uint32)
plt.imshow(final_image1)
plt.show()
```

```python
img_c=final_image+final_image1

plt.imshow(img_c)
plt.show()
```
